```python
from mido import MidiFile
```

```python
def ispart(track):
    count=0
    for msg in track:
        if not msg.is_meta:
            count+=1
            
    if count>100:
        return True
    else: return False
```

```python
import os
for root, dirs, files in os.walk('/home/mconati/ttmp/chopin/ChopinStyleTransfer/filtered/'):
    for file in files:
        name = root  + file
        mid = MidiFile(name, clip=True)
        count=0
        for track in mid.tracks:
            if ispart(track):
                count+=1
        if count>1:
            print(name)
```
